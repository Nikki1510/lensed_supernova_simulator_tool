```python
import numpy as np
from astropy.cosmology import Planck18
import matplotlib.pyplot as plt
```

```python
def pdf_red(zmax):
    z = np.linspace(0, zmax, 100)
    rate = 2.35 * 10**4  # from Perley 2020
    volume = Planck18.comoving_volume(z).to("Gpc**3").value
    z_rate = volume * rate
    rates = np.diff(z_rate)
    pdf = rates/np.nansum(rates)
    x = (z[:-1]+z[1:])*.5
    return x, pdf
```

```python
z, rate = pdf_red(zmax=5)

plt.figure(figsize=(6,3))
plt.plot(z, rate)
plt.xlabel("redshift")
plt.ylabel("PDF")
plt.savefig("../results/figures/unlensed_SN_rates.png", dpi=200)
```
