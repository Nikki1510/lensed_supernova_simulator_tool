```python
import numpy as np
from astropy.cosmology import Planck18
import matplotlib.pyplot as plt
```

```python
def rates(zmax):
    z = np.linspace(0, zmax, 100)
    rate = 2.35 * 10**4  # from Perley 2020
    volume = Planck18.comoving_volume(z).to("Gpc**3").value
    z_rate = volume * rate
    rates = np.diff(z_rate)
    pdf = rates/np.nansum(rates)
    x = (z[:-1]+z[1:])*.5
    return x, pdf

def rates_sanchez(zmax):
    z = np.linspace(0, zmax, 100)
    rate = 2.5 * 10**-5 * (1 + z)**1.5
    volume = Planck18.comoving_volume(z).to("Mpc**3").value
    z_rate = volume * rate
    rates = np.diff(z_rate)
    pdf = rates/np.nansum(rates)
    x = (z[:-1]+z[1:])*.5
    return x, pdf
```

```python
z1, rate1 = rates(zmax=1.5)
z2, rate2 = rates_sanchez(zmax=1.5)

plt.figure(figsize=(6,3))
plt.plot(z1, rate1, label='Constant rate')
plt.plot(z2, rate2, label='Sanchez 2021')
plt.xlabel("redshift")
plt.ylabel("PDF")
plt.legend()
#plt.savefig("../results/figures/unlensed_SN_rates_Sanchez.png", dpi=200, bbox_inches='tight')
```
