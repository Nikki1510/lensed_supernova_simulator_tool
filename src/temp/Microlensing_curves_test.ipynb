```python
import numpy as np
import matplotlib.pyplot as plt
import light_curves as micro
plt.rc("font", family="serif")
plt.rc("text", usetex=True)
```

```python
times = [micro.get_curves(kappa=0.362, gamma=0.342, s=0.790, source_redshift=0.3, filter="i")[2], 
         micro.get_curves(kappa=0.362, gamma=0.342, s=0.790, source_redshift=0.6, filter="i")[2],
         micro.get_curves(kappa=0.362, gamma=0.342, s=0.790, source_redshift=0.9, filter="i")[2],
         micro.get_curves(kappa=0.956, gamma=0.952, s=0.443, source_redshift=0.3, filter="i")[2],
         micro.get_curves(kappa=0.956, gamma=0.952, s=0.443, source_redshift=0.6, filter="i")[2],
         micro.get_curves(kappa=0.956, gamma=0.952, s=0.443, source_redshift=0.9, filter="i")[2]]

macro_curves = [micro.get_curves(kappa=0.362, gamma=0.342, s=0.790, source_redshift=0.3, filter="i")[0], 
                micro.get_curves(kappa=0.362, gamma=0.342, s=0.790, source_redshift=0.6, filter="i")[0],
                micro.get_curves(kappa=0.362, gamma=0.342, s=0.790, source_redshift=0.9, filter="i")[0],
                micro.get_curves(kappa=0.956, gamma=0.952, s=0.443, source_redshift=0.3, filter="i")[0],
                micro.get_curves(kappa=0.956, gamma=0.952, s=0.443, source_redshift=0.6, filter="i")[0],
                micro.get_curves(kappa=0.956, gamma=0.952, s=0.443, source_redshift=0.9, filter="i")[0]]
```

```python
times_s = [micro.get_curves(kappa=0.362, gamma=0.342, s=0.443, source_redshift=0.5, filter="i")[2], 
         micro.get_curves(kappa=0.362, gamma=0.342, s=0.616, source_redshift=0.5, filter="i")[2],
         micro.get_curves(kappa=0.362, gamma=0.342, s=0.790, source_redshift=0.5, filter="i")[2],
         micro.get_curves(kappa=0.956, gamma=0.952, s=0.443, source_redshift=0.5, filter="i")[2],
         micro.get_curves(kappa=0.956, gamma=0.952, s=0.616, source_redshift=0.5, filter="i")[2],
         micro.get_curves(kappa=0.956, gamma=0.952, s=0.790, source_redshift=0.5, filter="i")[2]]

macro_curves_s = [micro.get_curves(kappa=0.362, gamma=0.342, s=0.443, source_redshift=0.5, filter="i")[0], 
                micro.get_curves(kappa=0.362, gamma=0.342, s=0.616, source_redshift=0.5, filter="i")[0],
                micro.get_curves(kappa=0.362, gamma=0.342, s=0.790, source_redshift=0.5, filter="i")[0],
                micro.get_curves(kappa=0.956, gamma=0.952, s=0.443, source_redshift=0.5, filter="i")[0],
                micro.get_curves(kappa=0.956, gamma=0.952, s=0.616, source_redshift=0.5, filter="i")[0],
                micro.get_curves(kappa=0.956, gamma=0.952, s=0.790, source_redshift=0.5, filter="i")[0]]
```

```python

N_runs = 100

colors = ["C0", "C0", "C0", "C2", "C2", "C2"]
kappa_range = [0.362, 0.362, 0.362, 0.956, 0.956, 0.956]
gamma_range = [0.342, 0.342, 0.342, 0.952, 0.952, 0.952]
s_range = [0.790, 0.790, 0.790, 0.443, 0.443, 0.443]
z_range = [0.3, 0.6, 0.9, 0.3, 0.6, 0.9]

fig, ax = plt.subplots(2, 3, figsize=(15, 10))
fig.subplots_adjust(hspace=0.7)
ax = ax.flatten()
ax[0].text(0.75, 1.2, r"Image far from the lens galaxy (low $\kappa, \gamma$, high $s$)", fontsize=22, transform=ax[0].transAxes)
ax[3].text(0.75, 1.2, r"Image close to the lens galaxy (high $\kappa, \gamma$, low $s$)", fontsize=22, transform=ax[3].transAxes)

for im in range(len(ax)):
    ax[im].set_title("z = " + str(z_range[im]), fontsize=18)
    ax[im].set_ylabel("Mag", fontsize=18)
    ax[im].set_xlabel("Time [days]", fontsize=18)
    ax[im].invert_yaxis()
    ax[im].set_ylim(28, 19)
    ax[im].set_xlim(0, 130)
    
    ax[im].plot(times[im], macro_curves[im], color='black', lw=4)
    #if im < 3:
    #    ax[im].plot(times[im + 3], macro_curves[im + 3], color='gray', lw=1)
    #else:
    #    ax[im].plot(times[im - 3], macro_curves[im - 3], color='gray', lw=1)
        
    for lc in range(N_runs):
        ax[im].plot(times[im], micro.get_curves(kappa=kappa_range[im], gamma=gamma_range[im], s=s_range[im], 
        source_redshift=z_range[im], filter="i")[1], lw=1, color=colors[im], alpha=0.3)
    
fig.savefig("figures/Microlensing_test4.pdf", bbox_inches='tight')

plt.show()
```
